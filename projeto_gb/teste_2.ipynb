```python
import requests
import access
import pandas as pd

def obter_token():
    CLIENT_ID = access.client_id
    CLIENT_SECRET = access.secret_id

    AUTH_URL = 'https://accounts.spotify.com/api/token'

    auth_response = requests.post(AUTH_URL, {
        'grant_type': 'client_credentials',
        'client_id': CLIENT_ID,
        'client_secret': CLIENT_SECRET,
    })

    auth_response_data = auth_response.json()
    return auth_response_data['access_token']


access_token = obter_token()

headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
 }

BASE_URL = 'https://api.spotify.com/v1/'

# Track ID from the URI
EP_FULL = 'search?query=data+hackers&type=show&limit=5'  
busca_response_ep_full = requests.get(BASE_URL + EP_FULL, headers=headers)

EP_ID = 'shows/1oMIHOXsrLFENAeM743g93/episodes?limit=5'
busca_response_ep_id = requests.get(BASE_URL + EP_ID, headers=headers)

if busca_response_ep_full.status_code == 200:
    print("entro aqui - ok 200")
    resultados_full = busca_response_ep_full.json()
    print(resultados_full)
  
else:
    print(f'Erro na requisição da API do Spotify. Código de status: {busca_response_ep_full.status_code}')
    print('Resposta da API:')
    print(busca_response_ep_full.text)

if busca_response_ep_id.status_code == 200:
    print("entro aqui - ok 200")
    resultados_ep_id = busca_response_ep_id.json()
    print(resultados_ep_id['items'])
  
else:
    print(f'Erro na requisição da API do Spotify. Código de status: {busca_response_ep_id.status_code}')
    print('Resposta da API:')
    print(busca_response_ep_id.text)


# Verifica se 'shows' e 'items' estão presentes no JSON
if 'shows' in resultados_ep_id and 'items' in resultados_ep_id['items']:
    # Obtém a lista de itens
    podcasts_data = resultados_ep_id['items']

    # Cria o DataFrame
    df_resultados_ep_id = pd.DataFrame(podcasts_data, columns=['name', 'description', 'id'])

    # Imprime o DataFrame
    print("entrou na fd de ep_id")
    print(df_resultados_ep_id)
else:
    print('Estrutura do JSON não contém informações de podcasts.')

    


```
